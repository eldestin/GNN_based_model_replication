## 子图sampling 适用模型的问题


## 为什么需要子图训练？
首先明确一下子图训练的目的，两篇文章都提到，对于传统的GCN， GAT，在面对大型图(社交网络，知识图谱等)，会出现下列问题：
1. oversmoothing -> 收敛至相似的值，无法区分
2. low training efficiency -> 硬件问题,oom
3. 随着层数的增加，GCN 计算复杂度指数型增长

## 使用最多的图卷积(GCN), 图注意力(GAT)
在Ripple walk 和 Blocking-based sampling 两种方法中，其针对的都是图神经网络这一类型的模型，其propagation rule 如下图所示:



GCN 的模型参数仅仅为 权重W, GAT 的模型参数则是多了一个 attention 计算的weight。在两种模型的实现中，图的feature 矩阵 $ F \in R^{N*d}$, 其中N 为 node 的个数，d 为dimension的维度， 是从外部输入的（比如Cora中提供的是每篇论文的embedding）

## 我们的模型与其区别所在
我们的模型属于常见的KGE 类型，类似于TransE等。为每一个entity (在图中表示的就是node) 创建一个独立的embedding向量，来得到一个可以学习的feature。然后再通过图卷积来迭代并更新特征，权重。而上述两种子图选取的方法就会出现比较多的问题：
1. 对于一个mini-batch 内的子图(Ripple walk),其entity 并不一样，对应的embedding 矩阵也会不同
2. 很难在保证label一定比例的情况下进行筛选
3. 在测试的时候需要保证node feature的结构信息至少被用过一次
